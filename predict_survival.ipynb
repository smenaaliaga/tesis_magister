```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_tesis.csv')
```

```python
data.head()
```

```python
data_y = data.loc[:, ['died', 'survival_days']].to_numpy()
data_y
```

```python
#List of tuples
aux = [(e1,e2) for e1,e2 in data_y]

#Structured array
new_data_y = np.array(aux, dtype=[('Status', '?'), ('Survival_in_days', '<f8')])
new_data_y
```

```python
data_x = data.iloc[:, 5:]
data_x.head()
```

# Get Dummies

```python
data_x = pd.get_dummies(data_x)
data_x.head()
```

# Data Training y Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_x, new_data_y, test_size=0.3, random_state=42)
```

```python
from sksurv.ensemble import RandomSurvivalForest
```

```python
rsf = RandomSurvivalForest(n_estimators=1000,
                           min_samples_split=10,
                           min_samples_leaf=15,
                           n_jobs=-1,
                           random_state=42)
rsf.fit(X_train, y_train)
```

```python
rsf.score(X_test, y_test)
```
